## Import Libraries

`NetworkX` is used to handle the structure and math necessary in an
interconnected map and `matplotlib` helps render this visually.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

We'll create a blank working space:

```python
gworld= nx.Graph()
```

And add the nodes one by one:

```python
#--Nodes--
gworld.add_nodes_from(["Andre", "Beverly", "Ed", "Garth", "Fernando","Carol", "Diane", "Heather", "Ike", "Jane"])
```

Add the connections starting with Diane who seems to be central to this
organization:

```python
#--Edges--
#--Diane-- 
conn_diane=["Carol","Fernando","Garth","Ed","Beverly","Andre"]
gworld.add_edges_from([("Diane", connection) for connection in conn_diane])
```

check how its going:

```python
nx.draw(gworld, with_labels=True)
plt.show()
```

So far, the main hexagon is created, but it is not oriented in the exact way as
the image. I'll add the edges

```python
#--Carol Edges--
gworld.add_edges_from([("Carol", "Andre"), ("Carol", "Fernando")])
```

```python
#-- Adding last pairs--
gworld.add_edges_from([("Carol", "Andre"), ("Carol", "Fernando"), 
("Andre", "Carol"), ("Andre", "Beverly"), 
("Beverly", "Andre"),("Beverly", "Ed"), 
("Ed", "Beverly"), ("Ed", "Garth"), ("Garth", "Ed"), 
("Garth", "Fernando"), ("Garth", "Heather"),
("Fernando", "Garth"), ("Fernando", "Carol"), ("Fernando", "Heather"),
("Heather", "Ike"),
("Ike", "Jane")])
```

```python
nx.draw(gworld, with_labels=True)
plt.show()
```

The graph is almost there, but we're missing the inner square. . .

```python
gworld.add_edges_from([("Beverly", "Garth"), ("Andre", "Fernando")])
nx.draw(gworld, with_labels=True)
plt.show()
```

Checking out layout options

```python
format_g= nx.fruchterman_reingold_layout(gworld)
nx.draw(gworld,format_g, with_labels=True)
plt.show()
```

```python
format_g= nx.spring_layout(gworld)
nx.draw(gworld,format_g, with_labels=True)
plt.show()
```

## Final graph Render

```python
nx.draw(gworld,format_g, with_labels=True, node_size=2000,node_color="lavender", node_shape="s", font_size=10)
plt.show()
```
